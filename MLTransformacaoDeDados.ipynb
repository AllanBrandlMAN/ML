Importes

```python
import pandas as pd
from sklearn.compose import make_column_transformer# 1 vou importar uma funcao capaz de fazer algumas modificacoes hotEnconder (ele faz a transformacao de apenas uma coluna) ( make_column_transformer essa funcao e capaz de fazer diversas transformacoes na base de dados)
from sklearn.preprocessing import OneHotEncoder#ja realizamos a importacao das duas funcoes
from sklearn.preprocessing import LabelEncoder
```

Lendo o Arquivo

```python
dados = pd.read_csv('/content/churnTransformacao.csv')
```

Printando meus dados da tabela

```python
dados
```

Gerando informacoes completas da minha tabela

```python
dados.info()
```

Removendo minha variavel alvo e salvando ela numa variavel y

```python
x = dados.drop('churn', axis=1)# deletamos a variavel alvo utilizando o .drop
y = dados['churn']#criamos uma variavel para receber a variavel also independente
```

Printando a tabela sem a variavel alvo

```python
x
```

Printando a variavel alvo sozinha

```python
y
```

```python
colunas = x.columns
```

```python
one_hot = make_column_transformer((
    OneHotEncoder(drop = 'if_binary'),
    ['pais']
),
    remainder = 'passthrough',
    sparse_threshold = 0)
```

```python
x = one_hot.fit_transform(x)
```

Criado a lista com as novas variaveis modificadas

```python
one_hot.get_feature_names_out(colunas)
```

Printando a nova tabela transformado a mesma somente em numeros

```python
pd.DataFrame(x, columns = one_hot.get_feature_names_out(colunas))
```

```python
label_encoder = LabelEncoder()
```

```python
y = label_encoder.fit_transform(y)
```
