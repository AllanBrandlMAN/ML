```python
import pandas as pd
from sklearn.compose import make_column_transformer# 1 vou importar uma funcao capaz de fazer algumas modificacoes hotEnconder (ele faz a transformacao de apenas uma coluna) ( make_column_transformer essa funcao e capaz de fazer diversas transformacoes na base de dados)
from sklearn.preprocessing import OneHotEncoder#ja realizamos a importacao das duas funcoes
from sklearn.preprocessing import LabelEncoder
```

```python
dados = pd.read_csv('/content/marketing_investimento.csv')
```

```python
dados
```

```python
x = dados.drop('aderencia_investimento', axis=1)# deletamos a variavel alvo utilizando o .drop
y = dados['aderencia_investimento']#criamos uma variavel para receber a variavel also independente
```

```python
x # ira me retornar todas as variaveis explicativas exceto a variavel alvo
```

```python
y # retorna a variavel alvo
```

```python
colunas = x.columns
```

```python
one_hot = make_column_transformer((
    OneHotEncoder(drop = 'if_binary'),
    ['estado_civil','escolaridade','inadimplencia','fez_emprestimo']
),
    remainder = 'passthrough',
    sparse_threshold = 0)
```

Preparando a Transformacao

```python
x = one_hot.fit_transform(x) # ele vai transformar as colunas e salva las na mesma variavel x
```

```python
one_hot.get_feature_names_out(colunas) # retornar as colunas com os nomes novos
```

```python
pd.DataFrame(x, columns = one_hot.get_feature_names_out(colunas))
```

```python
label_encoder = LabelEncoder()#criando uma variavel para funcao LabelEncoder
```

```python
y = label_encoder.fit_transform(y)#ele transformor da colecao para numerico ficando a seguinte forma 0-nao e 1- sim
```
